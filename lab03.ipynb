```python
def min(lst):
    if len(lst) == 0:
        return None  
    min_value = lst[0]  
    for i in lst:
        if i < min_value:
            min_value = i
    return min_value

minimum = min(lst)
print(minimum)
```

```python
def max(lst):
    if len(lst) == 0:
        return None  # Return None for an empty list
    max_value = lst[0]  # Initialize max_value with the first element
    for i in lst:
        if i > max_value:
            max_value = i
    return max_value

maximum = max(lst)
print(maximum)
```

```python
Range= maximum - minimum
print(Range)
```
